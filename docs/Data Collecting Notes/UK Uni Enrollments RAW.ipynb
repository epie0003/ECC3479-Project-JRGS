# UK Uni Enrollments RAW
Notebook for exploring raw UK university enrollment data.

```python
from pathlib import Path

import pandas as pd

from IPython.display import display



# Locate UK enrollment data folder whether notebook runs from workspace root or docs/

candidate_dirs = [

    Path.cwd() / "UniEnrollmentsUK",

    Path.cwd().parent / "UniEnrollmentsUK",

    Path("UniEnrollmentsUK"),

    Path("../UniEnrollmentsUK"),

]



uk_dir = next((path.resolve() for path in candidate_dirs if path.exists()), None)

if uk_dir is None:

    raise FileNotFoundError("Could not find the 'UniEnrollmentsUK' folder.")



csv_files = sorted(uk_dir.glob("*.csv"), key=lambda p: p.name.lower())

if not csv_files:

    raise FileNotFoundError(f"No CSV files found in: {uk_dir}")





def read_csv_robust(file_path: Path) -> pd.DataFrame:

    try:

        return pd.read_csv(file_path, low_memory=False)

    except Exception:

        return pd.read_csv(

            file_path,

            engine="python",

            sep=None,

            on_bad_lines="skip",

        )





uk_raw_tables = {}

uk_raw_combined_parts = []



for file_path in csv_files:

    df = read_csv_robust(file_path)

    table_name = file_path.stem.replace("-", "_").replace(" ", "_")

    uk_raw_tables[table_name] = df



    df_with_source = df.copy()

    df_with_source["source_file"] = file_path.name

    uk_raw_combined_parts.append(df_with_source)



uk_uni_enrollments_raw_df = pd.concat(uk_raw_combined_parts, ignore_index=True)



print(f"Loaded {len(csv_files)} UK files from: {uk_dir}")

print("Files:")

for path in csv_files:

    print(f"- {path.name}")



print(f"\nCombined rows: {len(uk_uni_enrollments_raw_df)}, Columns: {len(uk_uni_enrollments_raw_df.columns)}")

display(uk_uni_enrollments_raw_df.head(20))
```

```python
# Profile each UK file structure (columns + sample rows)

for file_path in csv_files:

    print("\n" + "=" * 80)

    print(f"File: {file_path.name}")

    file_df = uk_raw_tables[file_path.stem.replace('-', '_').replace(' ', '_')]

    print(f"Shape: {file_df.shape}")

    print("Columns:", list(file_df.columns))

    display(file_df.head(8))
```

```python
# Locate where the real data table starts in each file

for file_path in csv_files:

    key = file_path.stem.replace('-', '_').replace(' ', '_')

    df = uk_raw_tables[key]

    print("\n" + "=" * 80)

    print(file_path.name)



    preview_text = df.apply(lambda row: " | ".join(map(str, row.values)), axis=1)

    matches = preview_text[preview_text.str.contains("Subject of study|subject of study|CAH|Total", case=False, regex=True)]

    print("Candidate rows:", matches.index[:10].tolist())

    if len(matches.index) > 0:

        first_idx = int(matches.index[0])

        start = max(0, first_idx - 3)

        end = min(len(df), first_idx + 8)

        display(df.iloc[start:end])
```

```python
# Build 2019-style grouped tables per year from each UK file

from io import StringIO

import re



target_columns = [

    "England",

    "Wales",

    "Scotland",

    "Northern Ireland",

    "Other UK",

    "Total UK",

]





def clean_subject_label(value: str) -> str:

    text = str(value).strip()

    match = re.match(r"^\((\d+)\)\s*(.*)$", text)

    if match:

        return f"{int(match.group(1)):02d} {match.group(2).strip()}"

    return text





def parse_2019_style_file(file_path: Path) -> dict:

    lines = file_path.read_text(encoding="utf-8-sig").splitlines()



    year_line = next((line for line in lines if line.startswith("Academic year,")), None)

    year_value = "2019/20"

    if year_line:

        parts = [part.strip().strip('"') for part in year_line.split(",")]

        if len(parts) > 1 and parts[1]:

            year_value = parts[1]



    header_idx = next((i for i, line in enumerate(lines) if line.startswith('"Subject of study"')), None)

    if header_idx is None:

        raise ValueError(f"Could not locate 'Subject of study' header in {file_path.name}")



    table_text = "\n".join(lines[header_idx:])

    df = pd.read_csv(StringIO(table_text), dtype=str)



    df = df.rename(columns={"Subject of study": "Subject"})

    for col in df.columns:

        if col != "Subject":

            df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", "", regex=False), errors="coerce")



    year_table = df[["Subject"] + [col for col in target_columns if col in df.columns]].copy()

    for col in target_columns:

        if col not in year_table.columns:

            year_table[col] = pd.NA

    year_table = year_table[["Subject"] + target_columns]



    return {year_value: year_table}





def parse_long_file(file_path: Path) -> dict:

    raw = pd.read_csv(file_path, header=None, dtype=str)

    header_candidates = ["Subject Area", "CAH level 1"]

    header_idx = None

    for candidate in header_candidates:

        found = raw[raw.iloc[:, 0].astype(str).str.strip() == candidate]

        if not found.empty:

            header_idx = int(found.index[0])

            break



    if header_idx is None:

        raise ValueError(f"Could not locate long-table header in {file_path.name}")



    header_values = [str(value).strip() for value in raw.iloc[header_idx].tolist()]

    data = pd.read_csv(file_path, skiprows=header_idx + 1, header=None, names=header_values, dtype=str)

    data = data.dropna(how="all")



    subject_col = next((col for col in data.columns if "subject" in str(col).lower() or "cah level" in str(col).lower()), None)

    country_col = next((col for col in data.columns if "country" in str(col).lower()), None)

    sex_col = next((col for col in data.columns if str(col).lower().strip() == "sex"), None)

    year_col = next((col for col in data.columns if "academic year" in str(col).lower()), None)

    number_col = next((col for col in data.columns if str(col).lower().strip() == "number"), None)

    entrant_col = next((col for col in data.columns if "entrant marker" in str(col).lower() or "first year marker" in str(col).lower()), None)

    level_col = next((col for col in data.columns if "level of study" in str(col).lower()), None)

    mode_col = next((col for col in data.columns if "mode of study" in str(col).lower()), None)



    required = [subject_col, country_col, sex_col, year_col, number_col]

    if any(col is None for col in required):

        raise ValueError(f"Missing required columns in {file_path.name}. Found columns: {list(data.columns)}")



    filtered = data.copy()

    if entrant_col:

        filtered = filtered[filtered[entrant_col].astype(str).str.strip() == "All"]

    if level_col:

        filtered = filtered[filtered[level_col].astype(str).str.strip() == "All"]

    if mode_col:

        filtered = filtered[filtered[mode_col].astype(str).str.strip() == "All"]

    filtered = filtered[filtered[sex_col].astype(str).str.strip() == "Total"]



    filtered[subject_col] = filtered[subject_col].map(clean_subject_label)

    filtered[number_col] = pd.to_numeric(filtered[number_col].astype(str).str.replace(",", "", regex=False), errors="coerce")



    year_tables = {}

    for year_value, year_df in filtered.groupby(year_col):

        pivot = (

            year_df

            .pivot_table(index=subject_col, columns=country_col, values=number_col, aggfunc="sum")

            .reset_index()

            .rename(columns={subject_col: "Subject", "All": "Total UK"})

        )



        for col in target_columns:

            if col not in pivot.columns:

                pivot[col] = pd.NA



        pivot = pivot[["Subject"] + target_columns]

        pivot = pivot.sort_values("Subject").reset_index(drop=True)

        year_tables[str(year_value).strip()] = pivot



    return year_tables





uk_grouped_tables_by_year = {}



for file_path in csv_files:

    if file_path.name == "2019.csv":

        parsed = parse_2019_style_file(file_path)

    else:

        parsed = parse_long_file(file_path)



    for year_key, year_table in parsed.items():

        uk_grouped_tables_by_year[year_key] = year_table



ordered_years = sorted(uk_grouped_tables_by_year.keys())

uk_grouped_comparison_df = pd.concat(

    [

        uk_grouped_tables_by_year[year].assign(AcademicYear=year)

        for year in ordered_years

    ],

    ignore_index=True,

)



print(f"Built grouped tables for {len(ordered_years)} years:")

print(", ".join(ordered_years))



for year in ordered_years:

    print("\n" + "=" * 80)

    print(f"Academic year: {year}")

    display(uk_grouped_tables_by_year[year])
```

```python
# Export grouped UK enrollment outputs

output_candidates = [

    Path.cwd() / "data" / "clean" / "uk_grouped",

    Path.cwd().parent / "data" / "clean" / "uk_grouped",

]

uk_output_dir = next((path for path in output_candidates if path.parent.exists()), output_candidates[0])

uk_output_dir.mkdir(parents=True, exist_ok=True)



for year, table in uk_grouped_tables_by_year.items():

    safe_year = year.replace("/", "-")

    table.to_csv(uk_output_dir / f"UK_enrollments_grouped_{safe_year}.csv", index=False)



uk_grouped_comparison_df.to_csv(uk_output_dir / "UK_enrollments_grouped_comparison_all_years.csv", index=False)



print(f"Exported per-year files to: {uk_output_dir}")

print(f"Combined comparison file: {uk_output_dir / 'UK_enrollments_grouped_comparison_all_years.csv'}")
```

```python
# Add AUS-aligned categorykey to UK grouped tables

import re



uk_subjectcode_to_categorykey = {

    1: 6,   # Medicine and dentistry -> Health

    2: 6,   # Subjects allied to medicine -> Health

    3: 1,   # Biological and sport sciences -> Natural & Physical Science

    4: 9,   # Psychology -> Society & Culture

    5: 6,   # Veterinary sciences -> Health

    6: 5,   # Agriculture, food and related studies -> Environment & Related

    7: 1,   # Physical sciences -> Natural & Physical Science

    9: 1,   # Mathematical sciences -> Natural & Physical Science

    10: 3,  # Engineering and technology -> Engineering & Related Tech

    11: 2,  # Computing -> Information Technology

    13: 4,  # Architecture, building and planning -> Architecture & Building

    15: 9,  # Social sciences -> Society & Culture

    16: 9,  # Law -> Society & Culture

    17: 8,  # Business and management -> Management & Commerce

    19: 9,  # Language and area studies -> Society & Culture

    20: 9,  # Historical, philosophical and religious studies -> Society & Culture

    22: 7,  # Education and teaching -> Education

    23: 11, # Combined and general studies -> Others

    24: 9,  # Media, journalism and communications -> Society & Culture

    25: 10, # Design, and creative and performing arts -> Creative Arts

    26: 5,  # Geography, earth and environmental studies -> Environment & Related

}





def extract_subject_code(subject_value: str):

    text = str(subject_value).strip()

    if text.lower().startswith("total"):

        return "TOTAL"



    match = re.match(r"^(\d{1,2})\b", text)

    if match:

        return int(match.group(1))



    match = re.match(r"^\((\d{1,2})\)", text)

    if match:

        return int(match.group(1))



    return None





def map_categorykey(subject_value: str):

    code = extract_subject_code(subject_value)

    if code == "TOTAL":

        return 99

    if code in uk_subjectcode_to_categorykey:

        return uk_subjectcode_to_categorykey[code]

    return 11





uk_grouped_tables_with_categorykey_by_year = {}

for year, table in uk_grouped_tables_by_year.items():

    enriched = table.copy()

    enriched["categorykey"] = enriched["Subject"].map(map_categorykey).astype("Int64")

    enriched = enriched.sort_values(["categorykey", "Subject"]).reset_index(drop=True)

    uk_grouped_tables_with_categorykey_by_year[year] = enriched



uk_grouped_comparison_with_categorykey_df = pd.concat(

    [

        uk_grouped_tables_with_categorykey_by_year[year].assign(AcademicYear=year)

        for year in sorted(uk_grouped_tables_with_categorykey_by_year.keys())

    ],

    ignore_index=True,

)



print("Added 'categorykey' to UK grouped tables using AUS key system (1..11, 99).")

print("Preview: 2019/20")

if "2019/20" in uk_grouped_tables_with_categorykey_by_year:

    display(uk_grouped_tables_with_categorykey_by_year["2019/20"])

else:

    first_year = sorted(uk_grouped_tables_with_categorykey_by_year.keys())[0]

    display(uk_grouped_tables_with_categorykey_by_year[first_year])
```

```python
# Export UK grouped outputs with categorykey (separate files)

uk_output_with_key_dir = uk_output_dir / "with_categorykey"

uk_output_with_key_dir.mkdir(parents=True, exist_ok=True)



for year, table in uk_grouped_tables_with_categorykey_by_year.items():

    safe_year = year.replace("/", "-")

    table.to_csv(uk_output_with_key_dir / f"UK_enrollments_grouped_{safe_year}_with_categorykey.csv", index=False)



uk_grouped_comparison_with_categorykey_df.to_csv(

    uk_output_with_key_dir / "UK_enrollments_grouped_comparison_all_years_with_categorykey.csv",

    index=False,

)



print(f"Exported UK grouped files with categorykey to: {uk_output_with_key_dir}")
```

```python
# Add Category names from AUS key system (for readability)

categorykey_to_category = {

    1: "Natural & Physical Science",

    2: "Information Technology",

    3: "Engineering & Related Tech",

    4: "Architecture & Building",

    5: "Environment & Related",

    6: "Health",

    7: "Education",

    8: "Management & Commerce",

    9: "Society & Culture",

    10: "Creative Arts",

    11: "Others",

    99: "Total",

}



uk_grouped_tables_with_category_by_year = {}

for year, table in uk_grouped_tables_with_categorykey_by_year.items():

    enriched = table.copy()

    enriched["Category"] = enriched["categorykey"].map(categorykey_to_category)

    enriched = enriched[["categorykey", "Category", "Subject", "England", "Wales", "Scotland", "Northern Ireland", "Other UK", "Total UK"]]

    enriched = enriched.sort_values(["categorykey", "Subject"]).reset_index(drop=True)

    uk_grouped_tables_with_category_by_year[year] = enriched



uk_grouped_comparison_with_category_df = pd.concat(

    [

        uk_grouped_tables_with_category_by_year[year].assign(AcademicYear=year)

        for year in sorted(uk_grouped_tables_with_category_by_year.keys())

    ],

    ignore_index=True,

)



uk_output_with_category_dir = uk_output_with_key_dir / "with_category_name"

uk_output_with_category_dir.mkdir(parents=True, exist_ok=True)



for year, table in uk_grouped_tables_with_category_by_year.items():

    safe_year = year.replace("/", "-")

    table.to_csv(uk_output_with_category_dir / f"UK_enrollments_grouped_{safe_year}_with_category.csv", index=False)



uk_grouped_comparison_with_category_df.to_csv(

    uk_output_with_category_dir / "UK_enrollments_grouped_comparison_all_years_with_category.csv",

    index=False,

)



print(f"Exported UK grouped files with Category names to: {uk_output_with_category_dir}")

print("Preview: 2019/20")

if "2019/20" in uk_grouped_tables_with_category_by_year:

    display(uk_grouped_tables_with_category_by_year["2019/20"])

else:

    first_year = sorted(uk_grouped_tables_with_category_by_year.keys())[0]

    display(uk_grouped_tables_with_category_by_year[first_year])
```
