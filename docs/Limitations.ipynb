# Data Limitations

This notebook documents known limitations across the datasets used in this project. These should be referenced when interpreting results and stated explicitly in any final report.

## 1. Coverage and Time Window Mismatches

- **AUS Enrollments** cover 2016–2024; **AUS Funding** covers 2019–2026 (projected). The inner join used in the multivariate analysis restricts the usable panel to **2019–2024 (~66 rows)**, losing three years of pre-JRG enrollment variation (2016–2018).
- **UK Enrollments** use academic years (e.g., 2020/21), converted to start-year integers. This introduces a **~6-month calendar misalignment** relative to AUS data, which uses calendar years. Any year-on-year comparison between the two countries should account for this offset.
- The **2019–2024 overlap** between AUS and UK enrollment data excludes years prior to 2019 from direct bilateral comparisons.

## 2. Missing Data

- **UK 'Other UK' region:** 257 out of 282 records have missing values for the `Other UK` column — a known HESA data gap, not a collection error. These are excluded before aggregation and do not bias estimates for the main FOE categories, but the `Other UK` regional breakdown cannot be analysed.
- **UK Wales / Northern Ireland:** 6 missing records each across the full subject-year panel. Minor impact; filled with zero before aggregation.
- **Employment by Industry:** 7 missing quarterly observations concentrated around 2020–2021. The pattern is systematic (COVID-related survey disruption), not random. Any regression using this series should include a COVID dummy or exclude the affected quarters rather than imputing.
- **AUS Funding — FOE codes:** 2 FOE codes have no matching category key, meaning a small number of funding records cannot be attributed to a specific broad field and are excluded from category-level analysis.

## 3. Structural Break — Job-Ready Graduates (JRG) Policy

- The **Job-Ready Graduates reform commenced in 2021**, changing student contribution rates and Commonwealth funding weights by field. This introduces a **structural break** in both the funding and enrollment series at 2021.
- Any regression pooling pre- and post-2021 data without a break indicator will produce biased time-trend coefficients. A post-JRG dummy or interaction term is required.
- The funding dataset provides per-unit rates (not total expenditure), so observed changes after 2021 reflect both the **price change** (new contribution rates) and any **volume change** (enrollment response). These effects are not separately identified without additional data.

## 4. Aggregation and Category Mapping

- **UK subject-to-FOE mapping:** UK HESA data is at subject level (~24 subjects) and is aggregated to 10 broad FOE categories for comparability with AUS data. This mapping is approximate; some subjects span multiple FOE groups. Aggregation decisions affect which categories appear to grow or shrink.
- **AUS Enrollments — 'Total' row:** The raw CSV includes a `Total` aggregate row that must be removed before analysis to avoid double-counting. All notebooks handle this, but any ad-hoc query against the raw file will overstate enrollment levels.
- **AUS Funding — per-unit rates vs totals:** The funding file contains per-student contribution rates, not total dollar expenditure. Category-year totals in the multivariate notebook are constructed by multiplying rates by enrollment counts; this product inherits uncertainty from both sources.

## 5. Statistical / Analytical Limitations

- **Simpson's Paradox is present in both AUS and UK enrollment data.** Aggregate positive time trends in enrollment conceal negative within-category trends (e.g., Creative Arts, Management & Commerce in AUS; Creative Arts, Natural & Physical Sciences, Environment in UK). Pooled regressions without category fixed effects will produce misleading coefficients.
- **Spurious size correlation in the multivariate dataset.** The positive aggregate correlation between funding and enrollments is driven by category size — larger fields receive more funding and enroll more students. This is not a causal effect; within-category analysis is required.
- **Small N.** The merged AUS panel has ~66 observations (11 categories × 6 years). With category fixed effects, the within-category variation used for identification is limited to 6 time points per category, constraining the precision of estimated effects.
- **No counterfactual for AUS.** The UK data can serve as a difference-in-differences control (UK did not implement JRG), but the subject-mapping approximation and calendar misalignment mean the parallel-trends assumption should be tested rather than assumed.
