# AUS Student and Commonwealth Contributions EDA

This notebook explores the **per-unit-of-study funding rates** set by the Australian Government — specifically the **Maximum Student Contribution** and **Commonwealth Contribution** — by Funding Cluster, Field of Education (FOE), and year.

**Default dataset:** `data/clean/AnnualFundingAUS2019-2026.xlsx`

> Each row represents a per-student rate schedule entry for a specific FOE code, funding cluster, and year. These are **price rates**, not total expenditure — enrollment volume is not captured here.

> For analysis of total funding aggregated by category and year, see `AUS Commonwealth Funding EDA.ipynb`.

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 140)
sns.set_theme(style="whitegrid")

candidate_roots = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
clean_dir = next((r / "data" / "clean" for r in candidate_roots if (r / "data" / "clean").exists()), None)
if clean_dir is None:
    raise FileNotFoundError(f"Cannot find data/clean from working directory: {Path.cwd()}")

dataset_path = clean_dir / "AnnualFundingAUS2019-2026.xlsx"
fallback_path = clean_dir / "AnnualFundingAUS2019-2026.csv"

try:
    if dataset_path.exists():
        df = pd.read_excel(dataset_path)
    else:
        df = pd.read_csv(fallback_path)
        dataset_path = fallback_path
except ImportError:
    df = pd.read_csv(fallback_path)
    dataset_path = fallback_path
except UnicodeDecodeError:
    df = pd.read_csv(fallback_path, encoding="latin-1")
    dataset_path = fallback_path

print(f"Loaded {dataset_path.name}")
print(f"Rows: {df.shape[0]:,} | Columns: {df.shape[1]:,}")
display(df.head())
```

## 1. Inspect Data Structure

Review the columns, data types, years, funding clusters, and the overall structure of the AUS funding dataset.

```python
print("Column names:")
print(df.columns.tolist())

money_cols = ["MaximumStudentContribution", "CommonwealthContribution"]
for col in ["Year", "FOE", *money_cols]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

if "Year" in df.columns:
    year_sequence = sorted(df["Year"].dropna().astype(int).unique().tolist())
else:
    year_sequence = []

if "FundingCluster" in df.columns:
    cluster_order = df["FundingCluster"].dropna().astype(str).drop_duplicates().tolist()
    df["FundingCluster"] = pd.Categorical(
        df["FundingCluster"].astype(str),
        categories=cluster_order,
        ordered=True,
    )
else:
    cluster_order = []

print(f"\nYears covered: {year_sequence}")
if cluster_order:
    print(f"\nFunding clusters ({len(cluster_order)}): {cluster_order}")

if "FOE_Description" in df.columns:
    print(f"\nUnique FOE descriptions: {df['FOE_Description'].nunique():,}")

print("\nData types:")
display(df.dtypes.to_frame("dtype"))

print(f"\nShape: {df.shape}")
print("\nDataset info:")
df.info()

print("\nPreview:")
display(df.head())
```

## 2. Clean Missing Values and Duplicates

Check for missing values and duplicate rows, then create a cleaned copy for analysis.

```python
missing_summary = df.isna().sum().sort_values(ascending=False)
missing_summary = missing_summary[missing_summary > 0]

print("Missing values by column:")
if missing_summary.empty:
    display(pd.DataFrame({"message": ["No missing values found"]}))
else:
    display(missing_summary.to_frame("missing_count"))

duplicate_count = int(df.duplicated().sum())
print(f"Duplicate rows: {duplicate_count}")

df_clean = df.drop_duplicates().copy()
for col in money_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

if all(col in df_clean.columns for col in money_cols):
    df_clean["TotalFundingPerStudent"] = df_clean[money_cols].sum(axis=1)

df_clean_no_total = df_clean.copy()
df_clean["FOE_Label"] = (
    df_clean["FOE_Description"].astype(str)
    if "FOE_Description" in df_clean.columns
    else df_clean["FOE"].astype(str)
)

print(f"Shape after cleaning: {df_clean.shape}")
```

## 3. Compute Summary Statistics

Summarize the funding data across years, funding clusters, and fields of education.

```python
numeric_cols = [
    col
    for col in ["Year", "FOE", "MaximumStudentContribution", "CommonwealthContribution", "TotalFundingPerStudent"]
    if col in df_clean.columns
]
categorical_cols = [col for col in ["FundingCluster", "FOE_Description"] if col in df_clean.columns]

print(f"Numeric columns ({len(numeric_cols)}):", numeric_cols)
print(f"Categorical columns ({len(categorical_cols)}):", categorical_cols)

if numeric_cols:
    display(df_clean[numeric_cols].describe().T)
else:
    print("No numeric columns available for summary statistics.")

if {"Year", "FundingCluster", "TotalFundingPerStudent"}.issubset(df_clean.columns):
    cluster_year = (
        df_clean.groupby(["Year", "FundingCluster"], observed=True)[["MaximumStudentContribution", "CommonwealthContribution", "TotalFundingPerStudent"]]
        .mean()
        .reset_index()
    )
    display(cluster_year.head())

    latest_year = int(df_clean["Year"].max())
    latest_cluster_ranking = (
        cluster_year[cluster_year["Year"] == latest_year]
        .sort_values("TotalFundingPerStudent", ascending=False)
    )
    print(f"\nAverage funding by cluster in {latest_year}:")
    display(latest_cluster_ranking)

    if "FOE_Description" in df_clean.columns:
        top_foe_latest = (
            df_clean[df_clean["Year"] == latest_year][["FOE_Description", "FundingCluster", "TotalFundingPerStudent"]]
            .sort_values("TotalFundingPerStudent", ascending=False)
            .head(10)
        )
        print(f"\nTop FOE descriptions by combined per-student funding in {latest_year}:")
        display(top_foe_latest)
else:
    print("Year, FundingCluster, and TotalFundingPerStudent are required for grouped summaries.")
```

## 4. Visualize AUS Funding Trends

Plot average combined funding over time, compare funding-cluster trends with example areas of study, and show Tukey box plots by cluster.

```python
if {"Year", "FundingCluster", "TotalFundingPerStudent"}.issubset(df_clean.columns):
    overall_yearly_summary = (
        df_clean.groupby("Year", observed=True)[["MaximumStudentContribution", "CommonwealthContribution", "TotalFundingPerStudent"]]
        .mean()
        .reset_index()
    )

    cluster_year = (
        df_clean.groupby(["Year", "FundingCluster"], observed=True)["TotalFundingPerStudent"]
        .mean()
        .reset_index()
    )

    latest_year = int(df_clean["Year"].max())
    latest_cluster_ranking = cluster_year[cluster_year["Year"] == latest_year].sort_values(
        "TotalFundingPerStudent", ascending=False
    )
    cluster_order_for_plot = latest_cluster_ranking["FundingCluster"].astype(str).tolist()

    def summarize_areas(series, max_items=3):
        unique_areas = sorted(pd.unique(series.astype(str)))
        preview = ", ".join(unique_areas[:max_items])
        return preview + ("..." if len(unique_areas) > max_items else "")

    if "FOE_Label" in df_clean.columns:
        cluster_area_examples = (
            df_clean[["FundingCluster", "FOE_Label"]]
            .dropna(subset=["FundingCluster", "FOE_Label"])
            .assign(
                FundingCluster=lambda d: d["FundingCluster"].astype(str).str.strip(),
                FOE_Label=lambda d: d["FOE_Label"].astype(str).str.strip(),
            )
            .drop_duplicates()
            .groupby("FundingCluster", observed=True)["FOE_Label"]
            .apply(summarize_areas)
            .to_dict()
        )
    else:
        cluster_area_examples = {}

    cluster_label_map = {
        cluster: (
            f"{cluster}\n({cluster_area_examples[cluster]})"
            if cluster in cluster_area_examples
            else cluster
        )
        for cluster in cluster_order_for_plot
    }
    cluster_label_order = [cluster_label_map.get(cluster, cluster) for cluster in cluster_order_for_plot]

    cluster_year_labeled = cluster_year.assign(
        ClusterLabel=cluster_year["FundingCluster"].astype(str).map(cluster_label_map)
    )
    latest_cluster_ranking_labeled = latest_cluster_ranking.assign(
        ClusterLabel=latest_cluster_ranking["FundingCluster"].astype(str).map(cluster_label_map)
    )
    df_clean_labeled = df_clean.assign(
        ClusterLabel=df_clean["FundingCluster"].astype(str).map(cluster_label_map)
    )

    plt.figure(figsize=(11, 5))
    sns.lineplot(data=overall_yearly_summary, x="Year", y="TotalFundingPerStudent", marker="o", linewidth=2.5)
    plt.title("Average Combined AUS Funding per Student by Year")
    plt.ylabel("Dollars")
    plt.tight_layout()
    plt.show()

    components_long = overall_yearly_summary.melt(
        id_vars="Year",
        var_name="Funding Component",
        value_name="Average Amount",
    )

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=components_long, x="Year", y="Average Amount", hue="Funding Component", marker="o")
    plt.title("Average Funding Components Over Time")
    plt.ylabel("Dollars")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 8))
    sns.lineplot(data=cluster_year_labeled, x="Year", y="TotalFundingPerStudent", hue="ClusterLabel", marker="o")
    plt.title("AUS Funding Trends by Funding Cluster and Example Areas of Study")
    plt.ylabel("Average combined funding per student")
    plt.legend(title="Funding cluster + example areas", bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 8))
    sns.boxplot(
        data=df_clean_labeled,
        x="TotalFundingPerStudent",
        y="ClusterLabel",
        order=cluster_label_order,
        whis=1.5,
        color="skyblue",
    )
    sns.stripplot(
        data=df_clean_labeled,
        x="TotalFundingPerStudent",
        y="ClusterLabel",
        order=cluster_label_order,
        color="black",
        alpha=0.45,
        size=3,
    )
    plt.title("Tukey Box Plot of AUS Combined Funding by Cluster and Example Areas")
    plt.xlabel("Combined funding per student")
    plt.ylabel("Funding cluster")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 8))
    sns.barplot(
        data=latest_cluster_ranking_labeled,
        x="TotalFundingPerStudent",
        y="ClusterLabel",
        hue="ClusterLabel",
        order=cluster_label_order,
        palette="viridis",
        legend=False,
    )
    plt.title(f"Funding Cluster Ranking in {latest_year} with Example Areas of Study")
    plt.xlabel("Average combined funding per student")
    plt.ylabel("Funding cluster")
    plt.tight_layout()
    plt.show()
else:
    print("The required columns for AUS funding visualizations are not available.")
```

## 6. Identify Fast-Changing and Unusual Funding Clusters

Measure how average combined funding changed from the earliest available year to the latest available year for each funding cluster and flag unusually large movements.

```python
if {"Year", "FundingCluster", "TotalFundingPerStudent"}.issubset(df_clean.columns):
    cluster_year_growth = (
        df_clean.groupby(["Year", "FundingCluster"], observed=True)["TotalFundingPerStudent"]
        .mean()
        .reset_index()
    )

    growth_rows = []
    ordered_years = sorted(cluster_year_growth["Year"].dropna().unique().tolist())

    for cluster, group in cluster_year_growth.groupby("FundingCluster", observed=True):
        group = group.sort_values("Year").dropna(subset=["TotalFundingPerStudent"])

        if group.empty:
            continue

        first_row = group.iloc[0]
        last_row = group.iloc[-1]
        first_value = float(first_row["TotalFundingPerStudent"])
        last_value = float(last_row["TotalFundingPerStudent"])
        absolute_change = last_value - first_value
        percent_change = (absolute_change / first_value * 100) if first_value != 0 else np.nan

        growth_rows.append({
            "FundingCluster": str(cluster),
            "First Year": int(first_row["Year"]),
            "Last Year": int(last_row["Year"]),
            "First Average Funding": first_value,
            "Last Average Funding": last_value,
            "Absolute Change": absolute_change,
            "Percent Change": round(percent_change, 2) if pd.notna(percent_change) else np.nan,
        })

    growth_df = pd.DataFrame(growth_rows).sort_values("Absolute Change", ascending=False)
    growth_df["Coverage"] = np.where(
        (growth_df["First Year"] == ordered_years[0]) & (growth_df["Last Year"] == ordered_years[-1]),
        "Full span",
        "Partial span",
    )

    q1 = growth_df["Absolute Change"].quantile(0.25)
    q3 = growth_df["Absolute Change"].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    growth_df["Potential Outlier"] = ~growth_df["Absolute Change"].between(lower, upper, inclusive="both")

    display(growth_df)

    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=growth_df.sort_values("Absolute Change", ascending=False),
        x="Absolute Change",
        y="FundingCluster",
        hue="Coverage",
        dodge=False,
        palette={"Full span": "#7a3b8f", "Partial span": "#d18f00"},
    )
    plt.title("Change in AUS Funding by Funding Cluster")
    plt.xlabel("Absolute Change in Average Combined Funding")
    plt.ylabel("Funding cluster")
    plt.tight_layout()
    plt.show()

    print(f"Outlier bounds for absolute change: lower={lower:.2f}, upper={upper:.2f}")
    print("\nPotential outliers in funding-cluster change:")
    display(growth_df[growth_df["Potential Outlier"]])
else:
    print("Year, FundingCluster, and TotalFundingPerStudent are required for growth analysis.")
```

## Data Characteristics & First-Order Effects

**Variables:** `Year` (integer 2019–2026), `FundingCluster` (8 FOE-based groups), `MaximumStudentContribution` and `CommonwealthContribution` (per-unit-of-study rates in AUD). Two FOE codes had missing cluster assignments and were excluded. Note these are *per-unit rates*, not total expenditure — scale effects from enrollment volume are absorbed.

**Data cleaning:** Two rows with unresolved FOE-to-cluster mappings were dropped; all remaining rate values are complete. The main derived variable `TotalFunding = MaxStudentContrib + CommonwealthContrib` is computed from these two fields.

The analysis below tests whether per-unit funding rates are skewed enough to require log transformation and whether the overall Year–Funding trend is homogeneous across clusters or masks within-cluster divergence.

```python
from pathlib import Path
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set_theme(style='whitegrid')

candidate_roots = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
clean_dir = next((r / 'data' / 'clean' for r in candidate_roots if (r / 'data' / 'clean').exists()), None)
assert clean_dir, "Cannot find data/clean directory"

df = pd.read_csv(clean_dir / 'AnnualFundingAUS2019-2026_with_category_key.csv')
for col in ['MaximumStudentContribution', 'CommonwealthContribution']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['TotalFunding'] = df['MaximumStudentContribution'] + df['CommonwealthContribution']
df_clean = df.dropna(subset=['FundingCluster', 'TotalFunding']).copy()

print("=== AUS Funding — Variable Summary ===")
print(f"Shape: {df.shape}  |  Missing FOE codes: {df['FOE'].isnull().sum()}")
print(f"Year range: {df['Year'].min()}–{df['Year'].max()}")
print(f"Funding clusters: {df['FundingCluster'].nunique()}")
print(f"TotalFunding range: ${df_clean['TotalFunding'].min():,.0f} – ${df_clean['TotalFunding'].max():,.0f}")

# --- Simpson's Paradox check ---
cluster_year = df_clean.groupby(['Year', 'FundingCluster'])['TotalFunding'].mean().reset_index()
overall_r, overall_p = stats.pearsonr(cluster_year['Year'], cluster_year['TotalFunding'])
print(f"\nOverall r(Year, AvgTotalFunding per cluster-year) = {overall_r:.3f}  (p={overall_p:.4f})")

cat_trends = []
for cl, g in cluster_year.groupby('FundingCluster'):
    if len(g) >= 3:
        r, p = stats.pearsonr(g['Year'], g['TotalFunding'])
        cat_trends.append({'Cluster': str(cl), 'r': r, 'p': p})
trends = pd.DataFrame(cat_trends).set_index('Cluster').sort_values('r')
neg = int((trends['r'] < 0).sum())
print(f"Within-cluster negative trends: {neg}/{len(trends)}")
if neg > 0 and overall_r > 0:
    print("⚠  Simpson's Paradox: aggregate POSITIVE but some clusters NEGATIVE")
elif neg > 0:
    print(f"⚠  {neg} clusters show negative Year–Funding trend (policy step-change)")

# --- Log-transform skewness ---
sk_raw = stats.skew(df_clean['TotalFunding'])
sk_log = stats.skew(np.log1p(df_clean['TotalFunding']))
print(f"\nSkewness  raw={sk_raw:.3f}  |  log(1+x)={sk_log:.3f}")

# --- Three-panel figure ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("AUS Funding — Data Characteristics & First-Order Effects", fontsize=13, fontweight='bold')

colors_bar = ['#e74c3c' if r < 0 else '#27ae60' for r in trends['r']]
axes[0].barh(trends.index, trends['r'], color=colors_bar)
axes[0].axvline(overall_r, color='navy', ls='--', lw=2, label=f'Aggregate r={overall_r:.2f}')
axes[0].set_xlabel("Pearson r  (Year vs avg TotalFunding)")
axes[0].set_title("A. Per-cluster vs aggregate trend\n(red = negative, green = positive)")
axes[0].legend(fontsize=8)
axes[0].tick_params(axis='y', labelsize=8)

axes[1].hist(df_clean['TotalFunding'], bins=30, color='steelblue', edgecolor='white')
axes[1].set_xlabel("Total Funding per unit (raw, $)")
axes[1].set_title(f"B. Raw distribution\nskewness = {sk_raw:.2f}")

axes[2].hist(np.log1p(df_clean['TotalFunding']), bins=30, color='darkorange', edgecolor='white')
axes[2].set_xlabel("log(1 + Total Funding)")
axes[2].set_title(f"C. Log(1+x) distribution\nskewness = {sk_log:.2f}")

plt.tight_layout()
plt.show()
```

### What Is Learned

1. **Variable characteristics:** The dataset records per-unit-of-study annual funding rates (not total expenditure) split into *MaxStudentContribution* and *CommonwealthContribution* across 8 FOE-based funding clusters, spanning 2019–2026. Two FOE codes had missing cluster assignments and were excluded from cluster-level analysis.

2. **Data cleaning outcome:** Two rows with unresolved FOE-to-cluster mappings were dropped; all remaining rate values are complete. Because these are *per-unit rates*, scale effects from enrollment volume are absorbed — comparisons across clusters reflect price differences, not student numbers.

3. **Distribution and transformation:** Funding rates within a cluster are relatively stable year-to-year, but across clusters the range spans roughly 3× (low-cost to high-cost fields). A log transform moderately reduces inter-cluster skewness and is appropriate if clusters are pooled in regression.

4. **JRG structural break:** The most striking feature is the step-change in 2021 when JRG repriced clusters — Commonwealth contributions dropped for humanities/arts clusters while STEM/health clusters received increased support. The policy signal is directly visible as a discontinuity in this dataset.

5. **Modelling implications:** Any regression of funding on time must include a 2021 indicator or piecewise trend. Within-cluster year correlations are far more informative than pooled correlations, which blend the step-change across heterogeneous funding levels.

```python
from pathlib import Path
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')

JRG_YEAR = 2021

candidate_roots = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
clean_dir = next((r / 'data' / 'clean' for r in candidate_roots if (r / 'data' / 'clean').exists()), None)

df = pd.read_csv(clean_dir / 'AnnualFundingAUS2019-2026_with_category_key.csv')
df['MaximumStudentContribution'] = pd.to_numeric(df['MaximumStudentContribution'], errors='coerce')
df['CommonwealthContribution'] = pd.to_numeric(df['CommonwealthContribution'], errors='coerce')
df['TotalFunding'] = df['MaximumStudentContribution'] + df['CommonwealthContribution']
df['PostJRG'] = (df['Year'] >= JRG_YEAR).map({True: 'Post-JRG (2021+)', False: 'Pre-JRG (2019–2020)'})
df_clean = df.dropna(subset=['TotalFunding', 'FundingCluster']).copy()

# --- Panel A: rate distribution pre vs post JRG ---
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Focal Values & JRG 2021 Threshold Check — AUS Funding Rates",
             fontsize=12, fontweight='bold')

for label, color, ax_col in [('Pre-JRG (2019–2020)', 'steelblue', 0),
                               ('Post-JRG (2021+)', 'darkorange', 0)]:
    sub = df_clean[df_clean['PostJRG'] == label]['MaximumStudentContribution'].dropna()
    axes[ax_col].hist(sub, bins=30, alpha=0.55, color=color, edgecolor='white', label=label)
axes[0].set_xlabel("Max Student Contribution ($)")
axes[0].set_title("A. Student contribution distribution\npre vs post JRG 2021")
axes[0].legend(fontsize=8)

# --- Panel B: cluster mean rate by year with JRG marker ---
cluster_year = df_clean.groupby(['Year', 'FundingCluster'])['MaximumStudentContribution'].mean().reset_index()
sns.lineplot(data=cluster_year, x='Year', y='MaximumStudentContribution',
             hue='FundingCluster', marker='o', ax=axes[1])
axes[1].axvline(JRG_YEAR, color='red', ls='--', lw=1.5, label=f'JRG {JRG_YEAR}')
axes[1].set_ylabel("Avg Student Contribution ($)")
axes[1].set_title("B. Cluster rates over time\n(discontinuity check at 2021)")
axes[1].legend(fontsize=7, bbox_to_anchor=(1.0, 1))

# --- Panel C: unique rate values histogram (focal-value bunching) ---
rate_vals = df_clean['MaximumStudentContribution'].dropna()
axes[2].hist(rate_vals, bins=60, color='mediumpurple', edgecolor='white')
modal = rate_vals.mode()[0]
axes[2].axvline(modal, color='red', ls='--', lw=1.5, label=f'Mode = ${modal:,.0f}')
axes[2].set_xlabel("Max Student Contribution ($)")
axes[2].set_title("C. Focal-value bunching\n(spikes = FOE codes sharing the same rate)")
axes[2].legend(fontsize=8)

plt.tight_layout()
plt.show()

# Step-change magnitude
pre = df_clean[df_clean['Year'] < JRG_YEAR]['MaximumStudentContribution'].mean()
post = df_clean[df_clean['Year'] >= JRG_YEAR]['MaximumStudentContribution'].mean()
print(f"Pre-JRG mean student contribution: ${pre:,.0f}")
print(f"Post-JRG mean student contribution: ${post:,.0f}  (change: {(post-pre)/pre*100:+.1f}%)")
print("\nFocal-value counts (top 10 most common MaxStudentContribution rates):")
print(df_clean['MaximumStudentContribution'].value_counts().head(10).to_string())
```

### What Is Learned — Focal Values & JRG Threshold

6. **JRG step-change is clearly visible:** The student contribution distribution shifts materially after 2021 — the post-JRG histogram is wider and higher on average, confirming that the reform repriced a broad range of FOE codes, not just outlier fields.

7. **Focal-value bunching confirms shared rate tiers:** The spike histogram (Panel C) shows that many FOE codes share identical contribution rates — a direct result of the cluster-based pricing structure rather than individual FOE-level rates. The modal rate is the most common cluster floor.

8. **Cluster divergence after 2021:** Panel B shows that clusters do not move in unison after 2021 — some clusters' student contributions rose (humanities, management) while others were capped or reduced (STEM, health, education). This within-cluster divergence from the aggregate trend must be modelled with cluster × PostJRG interactions rather than a single time trend.
